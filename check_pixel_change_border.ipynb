```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os 
from PIL import Image
directory='/home/nadee/Documents/thesis/binary/Deep/manual/'
save='/home/nadee/Documents/thesis/binary/masks/'
filenames = os.listdir(directory)
```

```python

def get_rid_of_boundary(mask):
    mask[0] = mask[1]
    mask[-1] = mask[-2]
    mask[:, 0] = mask[:,1]
    mask[:, -1] = mask[:,-2]
    return mask

def fileList(directory):
    "import files"
    image_files=[]

    for path,dirs,files in os.walk(directory):
        for filename in files:
           image_files.append(os.path.join(path,filename))
    return image_files

def grayScale(image):
    "use opencv to open images, returns directly as array"
    im = cv2.imread(image)
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    return gray
```

```python
imageFiles=fileList(directory)
print(len(imageFiles))
#print(filenames)
```

```python
im = grayScale(imageFiles[4])
plt.imshow(im)
plt.show()
```

```python
t = im
a = t.reshape(1, -1).squeeze()
n = np.array(a)
u = []
for i in n:
  if i not in u:
    u.append(i)
print(sorted(u))
```

```python
for i in range(len(imageFiles)):
    gray=grayScale(imageFiles[i])
    
    #Isolate the different classes
    nodata=np.where(gray==255)
    coral=np.where(gray==80) #1
    waves=np.where(gray==127)#5

    #give classes the correct label
    gray[nodata]=(1)
    gray[coral]=(0) #80
    gray[waves]=(1)#127

    #replace boundary
    final=get_rid_of_boundary(gray)
    #save
    cv2.imwrite('/home/nadee/Documents/thesis/binary/Deep/trainAnnotatedeep/' + filenames[i][:-14] +'.png', final) #png

print('save;')
```
